```python
import numpy as np
from sklearn import preprocessing
```

```python
x = np.array([[1., -1., 2.],
             [2., 0., 0.],
             [0., 1., -1.]])
```

```python
x_scaled = preprocessing.scale(x)
x_scaled
```

```python
x_scaled.mean(axis = 0)
```

```python
x_scaled.std(axis = 0)
```

```python
scaler = preprocessing.StandardScaler()
scaler.fit(x)
```

```python
scaler.transform(x)
```

```python
scaler.transform([[1,1,0]])
```

```python
from sklearn import datasets
import pandas as pd
wine = datasets.load_wine()
```

```python
wine_train = wine['data']
column = wine['feature_names']
```

```python
wine_train_pd = pd.DataFrame(wine_train,columns = column)
```

```python
wine_train_pd.head()
```

### .scale

```python
wine_scaled = pd.DataFrame(preprocessing.scale(wine_train),columns = column)
wine_scaled
```

### Standard Scaler

```python
scaler_wine = preprocessing.StandardScaler()
scaler_wine.fit(wine_train)
pd.DataFrame(scaler_wine.transform(wine_train),columns = column)
```

### MinMax Scaler

```python
min_max_scaler_object = preprocessing.MinMaxScaler()
min_max_scaler_object.fit(wine_train)
```

```python
pd.DataFrame(min_max_scaler_object.transform(wine_train),columns = column)
```
