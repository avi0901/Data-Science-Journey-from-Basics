```python
import numpy as np
```

```python
def data_conv():
    data_train1 = np.genfromtxt("Datasets/project_training_boston_x_y_train.csv",delimiter=',')
    y_train = data_train1[:,-1]
    x_train = data_train1[:,:-1]
    #x_train[:,-1] = 1    //Don't do this kind of thing because it do shallow copy
    M = len(x_train)
    new_col = np.ones((M,1))
    x_train = np.append(x_train,new_col,axis = 1)
    
    x_test = np.genfromtxt("Datasets/project_test_boston_x_test.csv",delimiter=',')
    # x_test does not have Y column in it so we have to add column containing 1 explicitly
    x_test = np.array(x_test)
    M = len(x_test)
    new_col = np.ones((M,1))
    x_test = np.append(x_test,new_col,axis = 1)
    return x_train, y_train, x_test
```

```python
def step_gradient(m,X_train,Y_train,learning_rate):
    N = len(X_train[0])
    M = len(X_train)
    m_slope=np.zeros(N)
    for i in range(M):
        X = X_train[i]
        Y = Y_train[i]
        for j in range(N):
            m_slope[j] += (-2/M)* (Y - (m*X).sum()) * (X[j])
    new_m = m - learning_rate * m_slope
    return new_m
```

```python
def cost(m, x, y):
    cost=0
    for i in range(len(x)):
        cost+=(1/len(x))*((y[i]-sum(m*x[i]))**2)
    print(cost)
```

```python
def gd(X_train, Y_train, learning_rate, num_iteration):
    M = len(X_train)
    N = len(X_train[0])
    m = np.zeros(N)
    for i in range(num_iteration):
        m = step_gradient(m,X_train,Y_train,learning_rate)
        print("itr= ", i, "cost=", end=' ')
        cost(m, X_train, Y_train)
    return m
```

```python
# Main code
X_train , Y_train, X_test = data_conv()
learning_rate = 0.1
num_iteration = 300
m = gd(X_train, Y_train, learning_rate, num_iteration)
Y_predict = (m*X_test).sum(axis = 1)
Y_predict = Y_predict.reshape(-1,1)
Y_predict
```

## Using Feature Scaling

```python
from sklearn import preprocessing
```

```python
def data_conv():
    data_train1 = np.genfromtxt("Datasets/project_training_boston_x_y_train.csv",delimiter=',')
    scaler = preprocessing.StandardScaler()
    y_train = data_train1[:,-1]
    x_train = data_train1[:,:-1]
    #x_train[:,-1] = 1    //Don't do this kind of thing because it do shallow copy
    square = []
    for i in x_train:
        square.append(i**2)
    square = np.array(square)
    x_train = np.append(x_train, square, axis = 1)
    M = len(x_train)
    new_col = np.ones((M,1))
    x_train = np.append(x_train,new_col,axis = 1)
    
    x_test = np.genfromtxt("Datasets/project_test_boston_x_test.csv",delimiter=',')
    # x_test does not have Y column in it so we have to add column containing 1 explicitly
    x_test = np.array(x_test)
    square = []
    for i in x_test:
        square.append(i**2)
    square = np.array(square)
    x_test = np.append(x_test, square, axis = 1)
    
    M = len(x_test)
    new_col = np.ones((M,1))
    x_test = np.append(x_test,new_col,axis = 1)
    
    
    scaler.fit(x_train)
    x_train = scaler.transform(x_train)
    x_test = scaler.transform(x_test)
    return x_train, y_train, x_test
```

```python
# Main code
X_train , Y_train, X_test = data_conv()
learning_rate = 0.11
num_iteration = 300
m = gd(X_train, Y_train, learning_rate, num_iteration)
Y_predict = (m*X_test).sum(axis = 1)
Y_predict = Y_predict.reshape(-1,1)
Y_predict
```
