## Open CV

```python
import cv2
```

```python
image = cv2.imread("Datasets/avi.jpg")
```

```python
image.shape
```

```python
window_name = 'Avinash'
cv2.imshow(window_name,image)
#waits for user to press any key 
#(this is necessary to avoid Python kernel form crashing)
cv2.waitKey(0) 
  
#closing all open windows 
cv2.destroyAllWindows() 
```

```python
# Extracting the height and width of an image
h, w = image.shape[:2]
# Displaying the height and width
print("Height = {},  Width = {}".format(h, w))
```

```python
image[1000][1000]
```

```python
(b, g, r) =  image[1000][1000]
print("R = {}, G = {}, B = {}".format(r, g, b))
```

```python
# We will calculate the region of interest 
# by slicing the pixels of the image
roi = image[1000 : 2000, 900 : 1600]
```

```python
window_name = 'Avinash'
cv2.imshow(window_name,roi)
#waits for user to press any key 
#(this is necessary to avoid Python kernel form crashing)
cv2.waitKey(0) 
#closing all open windows 
cv2.destroyAllWindows() 
```

```python
# resize() function takes 2 parameters, 
# the image and the dimensions
resize = cv2.resize(image, (1500, 800))
```

```python
window_name = 'Avinash'
cv2.imshow(window_name,resize)
cv2.waitKey(0) 
cv2.destroyAllWindows() 
```

The problem with this approach is that the aspect ratio of the image is not maintained. So we need to do some extra work in order to maintain a proper aspect ratio.

```python
ratio = 800/w
dim = (800, int(ratio*h))
resize_with_ar = cv2.resize(image,dim)
```

```python
window_name = 'Avinash'
cv2.imshow(window_name,resize_with_ar)
cv2.waitKey(0) 
cv2.destroyAllWindows() 
```
