## Data Cleaning

```python
documents = []
for category in movie_reviews.categories():
    for fileid in movie_reviews.fileids(category):
        documents.append((movie_reviews.words(fileid), category))
        
documents[0:5]
```

```python
import random
random.shuffle(documents)
documents[0:5]
```

Now we have all the words. Now i want to clean my all the files. Now we will remove stopwords and we will do lemmitization.

```python
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
```

```python
from nltk.corpus import stopwords
import string
stops = set(stopwords.words('english'))
punctuations = list(string.punctuation)
stops.update(punctuations)
```

```python
from nltk.corpus import wordnet
def get_simple_pos(tag):
    if tag.startswith('J'):
        return wordnet.ADJ
    elif tag.startswith('V'):
        return wordnet.VERB
    elif tag.startswith('N'):
        return wordnet.NOUN
    elif tag.startswith('R'):
        return wordnet.ADV
    else:
        return wordnet.NOUN
```

```python
from nltk import pos_tag
def clean_review(words):
    output_words = []
    for w in words:
        if w.lower() not in stops:
            pos = pos_tag([w])
            clean_word = lemmatizer.lemmatize(w, pos = get_simple_pos(pos[0][1]))
            output_words.append(clean_word.lower())
    return output_words
```

```python
documents = [(clean_review(doc),category) for doc, category in documents]
```

```python
documents[0]
```

## Building Features

```python
training_documents = documents[0:1500]
testing_documents = documents[1500:]
# We have alread randomly shuffeled the documents thats why we can do this
```

```python
#Rough explaination
a= [1,2]
b = [5,8]
a += b
a
```

```python
all_words = []
for doc in training_documents:
    all_words += doc[0]
```

```python
import nltk
```

```python
freq = nltk.FreqDist(all_words)
freq.most_common(15)
```

```python
common = freq.most_common(3000)
features = [i[0] for i in common]
features
```

```python
def get_feature_dict(words):
    current_features = {}
    words_set = set(words)
    for w in features:
        current_features[w] = w in words_set
    return current_features
```

```python
# Testing get_feature_dict function
get_feature_dict(training_documents[0][0])
# training_documents[0][0] #It contains words of document
```

```python
training_data = [(get_feature_dict(doc), category) for doc, category in training_documents]
testing_data = [(get_feature_dict(doc), category) for doc, category in testing_documents]
```
