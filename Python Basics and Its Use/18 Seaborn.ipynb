## Visualizing Statistical Relationships (ScatterPlot)

```python
df = sns.load_dataset('tips')
df.head()
```

```python
sns.relplot(x='total_bill',y='tip',data = df)
plt.plot()
```

```python
sns.relplot(x='total_bill',y='tip',data = df,kind='scatter')
plt.plot()
```

```python
sns.relplot(x='total_bill',y='tip',data = df,kind='line')
plt.plot()
```

```python
sns.relplot(x='total_bill',y='tip',data = df,hue='time')
plt.plot()
```

```python
sns.relplot(x='total_bill',y='tip',data = df,style='time')
plt.plot()
```

```python
sns.relplot(x='total_bill',y='tip',data = df,hue='time',style= 'smoker')
plt.plot()
```

```python
sns.relplot(x='total_bill',y='tip',data = df,hue='time',col= 'smoker')
plt.plot()
```

```python
sns.scatterplot(x='total_bill',y='tip',data = df,hue='time')
plt.plot()
```

## Visualizing Categorical Data

## Categorical scatter plot
- stripplot()
- swarmplot()

### stripplot()

```python
df = sns.load_dataset('tips')
df.head()
```

```python
#By default it is stripplot() kind = 'strip'
sns.catplot(x='sex',y='total_bill',data= df)
plt.plot()
```

```python
sns.catplot(x='sex',y='total_bill',data= df,jitter = 0.2)
plt.plot()
```

```python
sns.catplot(x='sex',y='total_bill',data= df,jitter = False)
plt.plot()
```

### swarmplot()

```python
sns.catplot(x='sex',y='total_bill',data= df,kind = 'swarm')
plt.plot()
```

```python
sns.catplot(x='sex',y='total_bill',data= df,kind = 'swarm',hue = 'smoker')
plt.plot()
```
