## LOGIN 

```python
USERNAME = 'SAMPLE_USERNAME'
PASSWORD = 'SAMPLE_PASSWORD'
```

```python
from selenium import webdriver
import time
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
```

```python
path = 'D:\Softwares\chromedriver'
driver = webdriver.Chrome(executable_path = path)

driver.maximize_window()
link = 'https://www.instagram.com/'
driver.get(link)

time.sleep(2)

username = driver.find_element_by_name('username')
password = driver.find_element_by_name('password')

username.send_keys(USERNAME)
password.send_keys(PASSWORD)

login = driver.find_element_by_xpath('//form[@id = "loginForm"]/div/div[3]/button')
login.submit()

wait = WebDriverWait(driver,20)
not_now = wait.until(EC.presence_of_element_located((By.XPATH,'//div[@class = "cmbtv"]/button')))
not_now.click()

not_now2 = wait.until(EC.presence_of_element_located((By.XPATH,'//button[contains(@class , "HoLwm")]')))
not_now2.click()
```

## Type for “food” in search bar and print all the names of the Instagram Handles that are displayed in list after typing “food”

```python
search_bar = wait.until(EC.presence_of_element_located((By.XPATH,'//input[contains(@class , "x3qfX")]')))
search_bar.send_keys('food')
time.sleep(5)
list_of_user = driver.find_elements_by_xpath('//div[contains(@class , "fuqBx")]/div/a/div/div[2]/div/div/div/div')
print("No of Names",len(list_of_user))
lis = []
print()
for i in list_of_user:
    if i.text.strip() != '':
        lis.append(i.text.strip())
        print(i.text.strip())
```

## Open the Instagram Handle of “So Delhi”

```python
search_bar = wait.until(EC.presence_of_element_located((By.XPATH,'//input[contains(@class , "x3qfX")]')))
search_bar.clear()
search_bar.send_keys('so delhi')
time.sleep(3)
first_selection = driver.find_element_by_xpath('//div[contains(@class , "fuqBx")]/div/a')
first_selection.click()
```

## Start following it. Print a message if you are already following

```python
follow = wait.until(EC.presence_of_element_located((By.XPATH,'//span[contains(@class , "bqE32")]/span/button')))
if follow.text.strip() == 'Follow':
    follow.click()
else:
    print('Already Following')
```

## After following, unfollow the instagram handle. Print a message if you have already unfollowed.

```python
follow = driver.find_element_by_xpath('//span[contains(@class , "bqE32")]/span/button')
if follow.text.strip() == 'Follow':
    print('Already Not Following')
else:
    follow.click()
    driver.find_element_by_xpath('//button[contains(@class , "Cab_")]').click()
```

## Liking first 30 posts

```python
driver.execute_script('window.scrollBy(0,5000);')

nextt = driver.find_element_by_xpath('//article[@class = "ySN3v"]/div/div/div/div/a/div')
for i in range(30):
    nextt.click()
    wait = WebDriverWait(driver,20)
    like = wait.until(EC.presence_of_element_located((By.XPATH,'//span[@class = "fr66n"]/button/div/span/*[name() = "svg"]')))
    
    #like = driver.find_element_by_xpath('//span[@class = "fr66n"]/button/div/span/*[name() = "svg"]')
    if like.get_attribute('aria-label') == 'Unlike':
        print('Already Liked')
    else:
        like.click()
    nextt = driver.find_element_by_xpath('//a[contains(@class , "_65Bje")]')
exitt = driver.find_element_by_xpath('//div[contains(@class , "yiMZG")]/button')
exitt.click()
driver.execute_script('window.scrollBy(0,0);')
```

## Unliking first 30 post 

```python
nextt = driver.find_element_by_xpath('//article[@class = "ySN3v"]/div/div/div/div/a/div')
for i in range(30):
    nextt.click()
    wait = WebDriverWait(driver,20)
    like = wait.until(EC.presence_of_element_located((By.XPATH,'//span[@class = "fr66n"]/button/div/span/*[name() = "svg"]')))
    if like.get_attribute('aria-label') == 'Unlike':
        like.click()
    else:
        print('Already UnLiked')
    nextt = driver.find_element_by_xpath('//a[contains(@class , "_65Bje")]')
exitt = driver.find_element_by_xpath('//div[contains(@class , "yiMZG")]/button')
exitt.click()
driver.execute_script('window.scrollBy(0,0);')
```
