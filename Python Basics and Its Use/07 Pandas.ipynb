```python
import pandas as pd
columns = ['sl','sw','pl','pw','flower_type']
iris = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data', names = columns)
arr =['Iris-setosa','Iris-versicolor','Iris-virginica']
for val in arr:
    a = iris[iris["flower_type"] == val]
    x = a.describe().values
    x_min = x[3]
    x_max = x[7]
    x_mean = x[1]
    y = [x_min, x_max, x_mean]
    for i in y:
        for p in i:
            print("{0:.2f}".format(p), end =" ")
        print(val)
```

#### Problem

Terror Attack City
Send Feedback

Given file "terrorismData.csv"
It is an open-source database including information on terrorist attacks around the world from 1970 through 2017. This dataset includes systematic data on domestic as well as international terrorist incidents that have occurred during this time period

Problem Statement :
The Most Dangerous city in Jammu and Kashmir and the terrorist group which is most active in that city?
Print count of number of attacks in that city as integer value.

Note:Ignoring the Unknown Terrorist Group.Here Dangerous related with the number of terrorist attacks.

Output Format :

City NumberOfAttack Group

```python
import pandas as pd
import numpy as np
terror = pd.read_csv("Datasets/terrorismData.csv")
Jk = terror[terror['State'] == 'Jammu and Kashmir']

cityarr = Jk["City"]
city,count = np.unique(cityarr,return_counts=True)
max_c = np.argmax(count)
mcity, mattack = city[max_c] , count[max_c]
Jk = Jk[Jk['Group'] != 'Unknown']
group1 = Jk[Jk["City"] == mcity]
group = group1["Group"]
grp, grpc = np.unique(group,return_counts=True)
max_g = np.argmax(grpc)
mgroup = grp[max_g]
print(mcity, mattack, mgroup)
#print(group.value_counts())
```

Given file "terrorismData.csv"

Problem Statement :
There was formation of new government in India on 26 May 2014. So current government's span is from 26th May 2014 to current. 

Find out two things from this period-
1. Total number of attacks done in this period in India. Find this count as integer.
2. Which Terrorist group was most active in this period in India. Most active means, group which has done maximum number of attacks. 
3.Ignore the Unknown group. 

Output Format :
TotalAttacks MostActiveTerroristGroup

```python
import pandas as pd
path="Datasets/terrorismData.csv"
df=pd.read_csv(path,encoding='ISO-8859-1')
df=df[df['Country']=='India']
df=df[df['Year']>=2014]
index_df1 = df[((df['Year'] == 2014) & (df['Month'] <5))].index
index_df2 = df[((df['Year'] == 2014) & (df['Month'] == 5) & (df['Day']<26))].index
df.drop(index_df1 , inplace = True)
df.drop(index_df2 , inplace = True)
attack = df.shape[0]
group = df['Group'].value_counts().index[1]
print(attack , group)
```

#### Problem

Terror DeadliestAttack
Send Feedback

Given file "terrorismData.csv"
It is an open-source database including information on terrorist attacks around the world from 1970 through 2017. This dataset includes systematic data on domestic as well as international terrorist incidents that have occurred during this time period

Problem Statement :
Most Deadliest attack in a history of HumanKind?
Print count of Killed people as integer value.

Note: Here Deadliest attack means, in which the most number of people killed.

Output Format :
NumberOfPeopleKilled Country TerroristGroup


```python
import pandas as pd
path="Datasets/terrorismData.csv"
df=pd.read_csv(path,encoding='ISO-8859-1')
df_killed = df['Killed']
df_country = df['Country']
df_group = df['Group']
index = df_killed.argmax()
print(int(df_killed[index]), df_country[index], df_group[index])
```

```python
import pandas as pd
path="Datasets/terrorismData.csv"
df=pd.read_csv(path,encoding='ISO-8859-1')
killed = max(df['Killed'])
country = df[df['Killed'] == killed]['Country'].values[0]
group = df[df['Killed'] == killed]['Group'].values[0]
print(int(killed),country,group)
```
