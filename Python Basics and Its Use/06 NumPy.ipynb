It is an open-source database including information on terrorist attacks around the world from 1970 through 2017. This dataset includes systematic data on domestic as well as international terrorist incidents that have occurred during this time period

Problem Statement :
Find the number of attack held between day 10 and day 20?(ignoring the year and month)(including both day)
Print count of NumberOFAttack as integer value.

Output Format :
count


```python
# Solution 1
import csv
import numpy as np
with open('Datasets/terrorismData.csv', errors= 'ignore') as file_obj:
    data = csv.DictReader(file_obj,skipinitialspace = True)
    
    attack = []
    for row in data :
        attack.append(row['Day'])
        
    attack_np = np.array(attack , dtype = float)
    
    array = np.logical_and(attack_np >= 10 , attack_np <= 20)
    ans  = attack_np[array]
    print(int(len(ans)))
```

#### Problem

Terror Casualty Attack

Send Feedback
Given file "terrorismData.csv"
It is an open-source database including information on terrorist attacks around the world from 1970 through 2017. This dataset includes systematic data on domestic as well as international terrorist incidents that have occurred during this time period

Problem Statement :
As we knew the Kargil ( in Jammu and Kashmir) War that took place between May 1999 and July 1999 (3 Months) ,so there was a huge conflict in Kashmir Valley during this period.
In this dataset, there is no information regarding the war between the two countries to find out the casualty during the war.
So find out the attack in this period in which maximum casualties happened.
Print the count of casualties (as integer), city in which that attack happened and name of attack group.

Note : Casualty = Killed + Wounded.Fill the empty value in killed or wounded feature to 0.

Output Format :
Casualty City TerroristGroup

```python
import csv
import numpy as np
with open("Datasets/terrorismData.csv",'r',encoding='ISO-8859-1') as file_obj:
    data = csv.DictReader(file_obj,skipinitialspace = True)
    
    details = []
    for row in data:
        details.append([row['Month'],row['Year'],row['City'],row['Group'],row['Wounded'],row['Killed'],row['State']])
        
    details_np = np.array(details)
    details_np = details_np[details_np[:,1]=='1999']
    details_np = details_np[(details_np[:,0]=='5') | (details_np[:,0]=='6') | (details_np[:,0]=='7')]
    details_np = details_np[details_np[:,6]=='Jammu and Kashmir']
        
    killed_np = details_np[:,5]
    killed_np[killed_np ==""] = '0'
    wounded_np = details_np[:,4]
    wounded_np[wounded_np ==''] = '0'
    killed_np = np.array(killed_np, dtype= float)
    wounded_np = np.array(wounded_np, dtype= float)
    casualty_np = killed_np + wounded_np
    casualty_np = np.array(casualty_np,dtype = int)
    
    index = casualty_np.argmax()
    
    print(casualty_np[index], details_np[index,2], details_np[index,3])
```

```python
import csv
import numpy as np

with open("Datasets/terrorismData.csv",'r',encoding='ISO-8859-1') as file:
    data = csv.DictReader(file, skipinitialspace = True)
    
    killed = []
    wounded = []
    city = []
    group = []
    
    for row in data:
        
        if row['State'] == "Jammu and Kashmir" and row["Year"] == "1999" and (row['Month']=="5" or row['Month']=="6" or row["Month"]=="7"):
            killed.append(row['Killed'])
            wounded.append(row['Wounded'])
            city.append(row['City'])
            group.append(row['Group'])
            
    killed = np.array(killed)
    killed[killed == ""] = "0.0"
    killed = np.array(killed, dtype=float)
    
    wounded = np.array(wounded)
    wounded[wounded == ""] = "0.0"
    wounded = np.array(wounded, dtype=float)
    
    casualties = killed + wounded
    index = casualties.argmax()
    
    print(int(casualties[index]), city[index], group[index] )
```

#### Problem

Terror City Casualty

Send Feedback
Given file "terrorismData.csv"
It is an open-source database including information on terrorist attacks around the world from 1970 through 2017. This dataset includes systematic data on domestic as well as international terrorist incidents that have occurred during this time period

Problem Statement :
Find top 5 Indian Cities which has most number of casualties ?
Print top 5 cities along with total casualties in that city. Print count of Casualty as integer value.

Note: Ignoring the City which is Unknown.
Casualty = Killed + Wounded.

Output Format :
city_1 casualty_1
city_2 casualty_2
city_3 casualty_3
city_4 casualty_4
city_5 casualty_5


```python
import csv
import numpy as np

city = []
casualty = []

with open("Datasets/terrorismData.csv", "r",encoding='ISO-8859-1') as file:
    data = csv.DictReader(file, skipinitialspace = True)
    for row in data:
        if row['Country']=='India':
            if row['City']!= 'Unknown':
                city.append(row['City'])
                casualty.append([row['Killed'],row['Wounded']])
casualty_np = np.array(casualty)
city_np = np.array(city)

casualty_np[casualty_np=='']=0
casualty_np=np.array(casualty_np,dtype=float)

#Sum alog row (axis = 1)
casualty_np=np.sum(casualty_np,axis=1)
def cityCasualty(cityarr,casualtyarr):
    arr = []
    unique = np.unique(cityarr)
    for i in unique:
        index, = np.where(cityarr==i)
        sumCasualty = casualtyarr[index].sum()
        arr.append([sumCasualty,i])
    arr.sort(reverse=True)
    return arr
arr = cityCasualty(city_np,casualty_np)
for i in range(5):
    print(arr[i][1],int(arr[i][0]))
```

#### Problem

Terrorism FrequentDay

Send Feedback
Given file "terrorismData.csv"
It is an open-source database including information on terrorist attacks around the world from 1970 through 2017. This dataset includes systematic data on domestic as well as international terrorist incidents that have occurred during this time period

Problem Statement :
Find the most frequent day of attack in a terrorismDataset ?

Note: Here np.unique can be used.

Print count of frequent day and number of attack as Integer value.

Output Format :
Day NumberOFAttack

```python
## Open and read data file as specified in the question
## Print the required output in given format
import csv
import numpy as np

with open("Datasets/terrorismData.csv", "r",encoding='ISO-8859-1') as file:
    data = csv.DictReader(file, skipinitialspace = True)
    day = []
    for row in data:
        day.append(row['Day'])
    day_np = np.array(day,dtype = int)
    
    unique_arr = np.unique(day_np)
    
    arr = []
    for i in unique_arr:
        temp = day_np[day_np== i]
        NoAttack = len(temp)
        arr.append([i,NoAttack])
        
    arr_np = np.array(arr)
    count = arr_np[:,1]
    ind = np.argmax(count)
    print(arr_np[ind][0],arr_np[ind][1])
```
