#### Problem

Plot Job_Year

Send Feedback
Given File 'amazonjobsdataset.csv'

It is a dataset including information on amazon job opening around the world from June 2011 to March 2018. This dataset is collected using Selenium and BeautifulSoup by scraping all of the jobs for Amazon job site.

Problem Statement :
Plot the line graph between no. of Job postings with respect to year.
Print the year and the number of job posting as integer value.
Note: Year should be in ascending order.

Output Format :

year1 job_posting1

year2 job_posting2
. . . 
. . .
. . .

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
data = pd.read_csv('Datasets/amazon_jobs_dataset.csv')
l=[]
for i in range(data.shape[0]):
    l.append(int(data["Posting_date"][i].split()[2]))
data["Year"]=np.array(l)
x = np.unique(data["Year"])
y=[]
for i in x:
    y.append(len(data[data["Year"]==i]))
plt.plot(x,y)
plt.show()
for i in range(len(x)):
    print(x[i],y[i])
```

##### Problem
Plot city_job
Send Feedback

Given File 'amazonjobsdataset.csv'
It is a dataset including information on amazon job opening around the world from June 2011 to March 2018. This dataset is collected using Selenium and BeautifulSoup by scraping all of the jobs for Amazon job site.

Problem Statement :
Plot the Pie chart between Indian cities vs No. of jobs opening.
Print the Indian cities and %age of Job distribution in India upto 2 decimal places.

Note: %age of Job distribution should be in descending order.

Output Format :

city1 percentage1

city2 percentage2

. . . 
. . .
. . .

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import csv

file_obj = open('Datasets/amazon_jobs_dataset.csv',encoding ='utf-8')
data = pd.read_csv(file_obj)
cities = data[data['location'].str.split(',',expand = True)[0] == 'IN']
cities = cities['location']
cities = cities.str.split(', ',expand = True)[2]
freq = cities.value_counts()
x= freq.index
y = freq
plt.pie(y,labels = x,autopct = "%.2f%%")
plt.show()

def per(i,sum):
    ans = (i*100)/sum
    return ans
for i in range(len(x)):
    print(x[i], format(per(y[i],sum(y)),'.2f'))
```

##### Problem

Plot Job_Java
Send Feedback

Given File 'amazonjobsdataset.csv'
It is a dataset including information on amazon job opening around the world from June 2011 to March 2018. This dataset is 
collected using Selenium and BeautifulSoup by scraping all of the jobs for Amazon job site.

Problem Statement :
Plot the scatter graph between year vs No. of jobs opening related to Java.
Print the year and number of Jobs opening in Java Profile.

Note: Use the Keyword 'Java' or 'java' in Basic Qualification feature for finding the job opening related to Java Profile. Print the year in ascending order.

Output Format :

year1 JobOpening1

year2 JobOpening2

. . . 
. . .
. . .

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import csv

file_obj = open('Datasets/amazon_jobs_dataset.csv',encoding ='utf-8')
file_data = csv.DictReader(file_obj,skipinitialspace = True)
lst = []
for row in file_data :
    bst = []
    if (('Java' in row['BASIC QUALIFICATIONS']) or ('java' in row['BASIC QUALIFICATIONS']) ):
        bst = row['Posting_date'].split(', ')
        lst.append(bst[1])
        

year = np.array(lst)
freq = np.unique(year, return_counts = True)[1]
years = np.unique(year, return_counts = True)[0]

plt.scatter(years,freq)
plt.show()

for i in range(len(freq)):
    print(years[i],freq[i])
```
