```python
import requests
from bs4 import BeautifulSoup
base_link = 'https://books.toscrape.com/catalogue/'
page_link = ['https://books.toscrape.com/catalogue/page-1.html','https://books.toscrape.com/catalogue/page-2.html']
book_links = []
for i in page_link:
    response = requests.get(i)
    data = BeautifulSoup(response.text,'html.parser')
    l1 = data.find_all(class_ = 'product_pod')
    for j in l1:
        link = j.h3.a['href']
        book_links.append(base_link+link)
book_links
```

```python
len(book_links)
```

```python
import re
book_details = []
for i in book_links:
    response = requests.get(i)
    data = BeautifulSoup(response.text,'html.parser')
    title = data.h1.string
    link = i
    qty = data.find(class_ = 'instock availability').contents[-1].strip()
    price = data.find(class_ = 'price_color').string
    qty = int(re.search('\d+',qty).group())
    price = float(re.search('[\d.]+',price).group())
    book_details.append([title,link,price,qty])
```

```python
import pandas as pd
df = pd.DataFrame(book_details,columns = ['Title', 'Link', 'Price', 'Quantity'])
df
```

#### Problem 3
Out of the first 250 titles with highest number of votes in 2018,find which title has the maximum duration.

```python
urls = ['https://www.imdb.com/search/title/?release_date=2018-01-01,2018-12-31&sort=num_votes,desc&start=1&ref_=adv_nxt']
i = 51 
base1 = 'https://www.imdb.com/search/title/?release_date=2018-01-01,2018-12-31&sort=num_votes,desc&start='
base2 = '&ref_=adv_nxt'
while i<=201 :
    url = base1 + str(i) + base2
    urls.append(url)
    i = i+50
urls
```

```python
titles = []
duration = []
for i in urls:
    response = requests.get(i)
    data = bs(response.text,'html.parser')
    movies = data.find_all(class_ = 'lister-item mode-advanced')
    for j in movies:
        titles.append(j.h3.a.string)
        runtime1 = j.find(class_ = 'runtime')
        if runtime1 is None :
            duration.append(0)
            continue
        runtime = runtime1.string
        runtime = re.search('\d+',runtime).group()
        duration.append(int(runtime))
print(len(titles))
print(len(duration))
```

```python
max_dur_ind = duration.argmax()
```

#### Problem 5
Find and print the src of the <img> tag which occupies the maximum area on the page.

```python
req = data.find_all('img')
source = []
area = []
for i in req:
    source.append(i['src'])
    h = i['height']
    if i['height'] is None:
        h = 0
    h = int(h)
    w = i['width']
    if i['width'] is None:
        w = 0
    w = int(w)
    a = h*w
    area.append(a)
```

```python
import numpy as np
area_np = np.array(area)
max_ind = area_np.argmax()
print(source[max_ind])
print(area[max_ind])
```

#### Problem 7
Find and print the names of all the different authors from all pages of this website
Note : Print the names of all authors line wise sorted in dictionary order


```python
link = 'http://quotes.toscrape.com'
pages = ['http://quotes.toscrape.com/']
current = 'http://quotes.toscrape.com/'
response = requests.get(current)

while response.status_code == 200:
    data = bs(response.text,'html.parser')
    nex = data.find(class_ = 'next')
    if nex is None:
        break
    nex_url = nex.a['href']
    pages.append(link+nex_url)
    current = link+nex_url
    response = requests.get(current)
```

```python
len(pages)
```

```python
authors = []
for i in pages:
    response = requests.get(i)
    data = bs(response.text,'html.parser')
    author = data.find_all(class_ = 'author')
    for j in author:
        name = j.string
        if name in authors:
            continue
        else:
            authors.append(name)
```

```python
authors.sort()
for i in authors:
    print(i)
```

#### Problem 8
Find the birth date of authors whose name start with 'J' from this website
Note : Print a dictionary containing the name as key and the birth date as value.The Names of authors should be alphabetically sorted.


```python
link = 'http://quotes.toscrape.com'
my_dict = {}
for i in pages:
    response = requests.get(i)
    data = bs(response.text,'html.parser')
    quote = data.find_all(class_ = 'quote')
    for j in quote:
        author = j.find(class_ = 'author').string
        if not(author.startswith('J')):
            continue
        if author in my_dict:
            continue
        about = link + j.a['href']
        response1 = requests.get(about)
        data2 = bs(response1.text,'html.parser')
        date = data2.find(class_ = 'author-born-date').string
        my_dict[author] = date
```

```python
print(my_dict)
```

```python
#print in sorted order
li = sorted(my_dict)  #sorted list of key
print('{',end="")
for i in li:
    ans = "''" + i + "' : '" + my_dict[i] + "'"
    print(ans,end=', ')
print('}')
```
