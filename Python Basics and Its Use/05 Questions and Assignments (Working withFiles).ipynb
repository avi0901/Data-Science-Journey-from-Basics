##### First 3 lines

Given file "year2017.csv"

It is an open-source database including information on terrorist attacks around the world from 1970 through 2017. This dataset includes systematic data on domestic as well as international terrorist incidents that have occurred during this time period

Problem Statement : 

Open and read the file

Print the first 3 rows of the file, excluding header (in different lines)

Print columns of one row, separated by space

Output Format:

col 1 col 2 col3 col_4.... 
col 1 col 2 col3 col_4.... 
col 1 col 2 col3 col_4....

```python
import csv
with open('Datasets/year2017.csv') as file_obj :
    file_data = csv.reader(file_obj)
    
    k = 0
    for row in file_data:
        if(k==0):
            k+=1
            continue
        elif k>=4:
            break
        for i in range(0,len(row)) :
            print(row[i], end=" ")
        print()
        k+=1
```

```python
import csv
with open('Datasets/year2017.csv') as file_obj :
    file_data = csv.reader(file_obj)
    file_list = list(file_data)
    
    for i in range(1,4):
        for j in range(len(file_list[0])):
            print(file_list[i][j],end=" ")
        print()
```

##### Casualties from Explosives

Given file "year2017.csv"

It is an open-source database including information on terrorist attacks around the world from 1970 through 2017. This dataset includes systematic data on domestic as well as international terrorist incidents that have occurred during this time period

Problem Statement : Find and print total numberof casualties with "Explosives" Weapon_type.

Print the count as integer value.

```python
import csv
with open('Datasets/year2017.csv') as file_obj :
    file_data = csv.DictReader(file_obj,skipinitialspace = True)
    file_list = list(file_data)
    casualties = []
    for row in file_list[1:]:
        if row['casualities'] != "" and row['Weapon_type']=="Explosives":
            casualties.append(float(row['casualities']))
print(int(sum(casualties)))
```

##### Month vs Killed

Given file "year2017.csv"

It is an open-source database including information on terrorist attacks around the world from 1970 through 2017. This dataset includes systematic data on domestic as well as international terrorist incidents that have occurred during this time period

Problem Statement : Find the total number of people killed in each month.

Print the month and count of killed people as integer value.

```python
import csv
with open('Datasets/year2017.csv') as file_obj :
    file_data = csv.DictReader(file_obj,skipinitialspace = True)
    file_list = list(file_data)
    
month_killed = {}
for row in file_list:
    key = row['Month']
    value = row['Killed']
    
    if value != '':
        value= int(float(value))
    else:
        value = 0
        
    if key in month_killed:
        month_killed[key] += value
    else:
        month_killed[key] = value
        
for i in month_killed:
    print(i, month_killed[i])
```

Given File 'amazonjobsdataset.csv'

It is a dataset including information on amazon job opening around the world from June 2011 to March 2018. This dataset is collected using Selenium and BeautifulSoup by scraping all of the jobs for Amazon job site.

Problem Statement:

Find the month having most job openings in Year 2018 ?

Print the month (Month Name ie January, February, March) and Number of job opening as Integer Value

Output Format:

MonthName Count

```python
import csv
with open('Datasets/amazon_jobs_dataset.csv',errors = 'ignore') as file_obj:
    file_data = csv.DictReader(file_obj,skipinitialspace = True)
    month_opening = {}
    
    for row in file_data :
        lst = []
        if '2018' in row['Posting_date']:
            lst = row['Posting_date'].split()
            month_opening[lst[0]] = month_opening.get(lst[0],0)+1
            
    ##print(month_opening)
    month = ''
    opening = 0
    for i in month_opening:
        if month_opening[i] > opening :
            opening = month_opening[i]
            month = i
    print(month , opening)
```

##### Job Language

Send Feedback

Given File 'amazonjobsdataset.csv'

It is a dataset including information on amazon job opening around the world from June 2011 to March 2018. This dataset is collected using Selenium and BeautifulSoup by scraping all of the jobs for Amazon job site.

Problem Statement :

Among Java, C++ and Python, which of the language has more job openings in India for Bachelor Degree Holder?

Print the Language(i.e Java.C++,Python) and number of job opening as integer value.

Note: Here we will use the BASIC QUALIFICATIONS feature to find out whether bachelor degree for Job is required or not. Keywords that can be used are 'Bachelor', 'BS' and 'BA' and we will use the BASIC QUALIFICATIONS feature to find out whether Language is required for the job or not. Keywords that is used for language searching are 'Java'. 'C++' or 'Python'. (There case should not be changed). 

Output Format:

Language Count

```python
import csv
with open('Datasets/amazon_jobs_dataset.csv',errors = 'ignore') as file_obj:
    file_data = csv.DictReader(file_obj,skipinitialspace = True)
    lang_count = {}
    
    for row in file_data :
        if 'IN' in row['location']:
            if 'Bachelor' in row['BASIC QUALIFICATIONS'] or 'BS' in row['BASIC QUALIFICATIONS'] or 'BA' in row['BASIC QUALIFICATIONS']:
                if 'Java' in row['BASIC QUALIFICATIONS']:
                    lang_count['Java'] = lang_count.get('Java',0)+1
                if 'C++' in row['BASIC QUALIFICATIONS']:
                    lang_count['C++'] = lang_count.get('C++',0)+1
                if 'Python' in row['BASIC QUALIFICATIONS']:
                    lang_count['Python'] = lang_count.get('Python',0)+1
                    
                
    #print(lang_count)
    lang = ''
    count = 0
    for i in lang_count:
        if lang_count[i] > count :
            count = lang_count[i]
            lang = i
    print(lang , count)
```

Given File 'amazonjobsdataset.csv'

Problem Statement :

Find the country does Amazon need the most number of Java Developer?

Note:Here we will use the BASIC QUALIFICATIONS feature to find out whether Java is required for the job or not.Keyword is used is 'Java'.(Here case should not be changed).

```python
import csv
with open('Datasets/amazon_jobs_dataset.csv',errors = 'ignore') as file_obj:
    file_data = csv.DictReader(file_obj,skipinitialspace = True)
    country_count = {}
    
    for row in file_data :
        lst = []
        if 'Java' in row['BASIC QUALIFICATIONS']:
            lst = row['location'].split(',')
            country_count[lst[0]] = country_count.get(lst[0],0)+1
            
    ##print(country_count)
    country = ''
    count = 0
    for i in country_count:
        if country_count[i] > count :
            count = country_count[i]
            country = i
    print(country , count)
```
